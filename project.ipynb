```python
import gensim
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from nltk.stem import WordNetLemmatizer, SnowballStemmer
from nltk.stem.porter import *
import numpy as np
np.random.seed(2018)
import nltk
nltk.download('wordnet')
from gensim import corpora, models
from gensim.models import LdaModel, LdaMulticore, CoherenceModel
import matplotlib.pyplot as plt
import pyLDAvis.gensim

#pre-process the data by using tokenization and stopword removal
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS:
            result.append(token)
    return result

processed_docs = []
for i in range(100):
	if (isinstance(data.loc[i, "abstract"], str)):
		processed_docs.append(preprocess(data.loc[i, "abstract"]))
#use bag of words to represent the words in the document
dictionary = gensim.corpora.Dictionary(processed_docs)
bow_corpus = [dictionary.doc2bow(doc) for doc in processed_docs]
#find the number of topics that maximizes the coherence value
coherence_values = []
lda_models = []
for i in range(10, 60, 10):
    #use LDA to find the topics that occur in a collection of documents
    lda_model = gensim.models.LdaMulticore(bow_corpus, num_topics = i, id2word = dictionary, passes = 2, workers = 2)
    lda_models.append(lda_model)
    #use the coherence value to evaluate the topic model
    lda_coherence_model = CoherenceModel(model = lda_model, texts = processed_docs, dictionary = dictionary, coherence = 'c_v')
    coherence_value = lda_coherence_model.get_coherence()
    coherence_values.append(coherence_value)
#plot the coherence values
x = list(range(10, 60, 10))
y = coherence_values
plt.plot(x, y)
plt.xlabel("Number of Topics")
plt.ylabel("Coherence Value")
plt.show()
lda_model = lda_models[coherence_values.index(max(coherence_values))]
print(max(coherence_values))
print(lda_model.print_topics())
#visualize the topics and the keywords
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lda_model, bow_corpus, dictionary)
vis
```

```python
import gensim
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from nltk.stem import WordNetLemmatizer, SnowballStemmer
from nltk.stem.porter import *
import numpy as np
np.random.seed(2018)
import nltk
nltk.download('wordnet')
from gensim import corpora, models
from gensim.models import LsiModel, CoherenceModel

#pre-process the data by using tokenization and stopword removal
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS:
            result.append(token)
    return result

processed_docs = []
for i in range(100):
	if (isinstance(data.loc[i, "abstract"], str)):
		processed_docs.append(preprocess(data.loc[i, "abstract"]))
#use bag of words to represent the words in the document
dictionary = gensim.corpora.Dictionary(processed_docs)
bow_corpus = [dictionary.doc2bow(doc) for doc in processed_docs]
#use LSI to find the topics that occur in a collection of documents
lsi_model = LsiModel(corpus = bow_corpus, id2word = dictionary, num_topics = 50, decay = 0.5)
#use the coherence value to evaluate the topic model
lsi_coherence_model = CoherenceModel(model = lsi_model, texts = processed_docs, dictionary = dictionary, coherence = 'c_v')
coherence_value = lsi_coherence_model.get_coherence()
print(coherence_value)
print(lsi_model.print_topics())
```
